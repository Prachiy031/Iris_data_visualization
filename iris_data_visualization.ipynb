```python
s_width = df['sepal_width']
s_length = df['sepal_length']
p_width = df['petal_width']
p_length = df['petal_length']
species = df['species']
```

```python
plt.scatter(s_length,p_length)    #scatter plot
```

```python
plt.scatter(s_width,p_width)
```

```python
plt.plot(s_width,species,'r--')            #plot graph
plt.xlabel('s_width')
plt.ylabel('species')
plt.title('relation between s_width and species')
```

```python
plt.hist(s_length,bins=10)  #histogram
plt.xlabel('s_length')
plt.ylabel('count')
plt.title('relation between s_length and count')
```

```python
values = df.select_dtypes(include='number')           #selecting only numerical data 
plt.boxplot(values,vert=True,patch_artist=True)
```

```python
# The diagonal elements in a pairplot show the histogram by default
# We can update these elements to show other things, such as a kde
sns.pairplot(df, hue="species", size=3, diag_kind="kde")
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df,size=4)
```
